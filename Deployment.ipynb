```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import word_tokenize
import re 

from sklearn.pipeline import make_pipeline
import preprocessing

import joblib
import flask
from flask import Flask, request, render_template
```

```python
model = joblib.load('tfidf_pipeline.pkl')
```

```python
app = flask.Flask(__name__, template_folder='templates')

@app.route('/', methods=['GET', 'POST'])
def index():
    mapping = {"EG":"Egypt", "LB": "Lebanon", "LY": "Libya", "MA": "Morocco", "SD":"Sindhi"}
    if flask.request.method == 'GET':
        return(flask.render_template('trial.html'))
    
    if flask.request.method == 'POST':
        tweet = [flask.request.form['uname']]
        text = [preprocessing.processing(x) for x in tweet]
        y_pred = model.predict(text)
        label = mapping[y_pred[0]]
        
        return render_template('trial.html', result=f'The predicted dialect is: {label}')
```

```python
app.run()
```
